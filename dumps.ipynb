```python
from kloppy import pff

from mplsoccer import Pitch
import matplotlib.pyplot as plt

from kloppy.domain import CoordinateSystem

dataset = pff.load_event(
        event_data="/home/jupiter/ufmg/thesis/gandula-expected-goals/data/01_raw/event_data/3812.json",
        meta_data="/home/jupiter/ufmg/thesis/gandula-expected-goals/data/01_raw/metadata/3812.json",
        roster_data="/home/jupiter/ufmg/thesis/gandula-expected-goals/data/01_raw/rosters/3812.json",
        coordinates="pff",
    )
```

```python
pitch = Pitch()
fig, ax = pitch.draw()

# Example: Plot all passes
for event in dataset.events:
    if event.event_type.name == "SHOT":
        x, y = event.coordinates.x, event.coordinates.y

        # Optional: differentiate goals from missed shots
        color = 'red' if getattr(event, 'outcome', None) and event.outcome.name == "GOAL" else 'blue'
        
        # Plot the shot as a scatter point (or use ax.annotate for arrow if you want shot direction)
        ax.scatter(x+10, y+10, color=color, edgecolors='white', linewidth=1.5, s=100, zorder=3)

plt.title('Shot Map')
plt.show()
```
